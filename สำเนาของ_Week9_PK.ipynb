```python
fig = px.scatter(df, ## data frame
                x='sugars', ## แกน X
                y='rating', ## แกน Y
                #hover_name='name',
                color = 'shelf', ## ปรับสีตาม shelf
                size = 'cups', ## ปรับขนาดตาม cups
                hover_data=['calories','carbo'], ## เมื่อเมาส์ไปวางจะ show อะไร
                hover_name= df['name'].astype(str)+" : "+df['mfr'].astype(str),
                title='Cereal_ratings VS Sugars')
fig.show()
```
