```python
import pandas as pd

file_id = "1R_oYvMigqRVQ8Qrq7QNoyxUk_zorr_I3"

url = f"https://drive.google.com/uc?export=download&id={file_id}"
print(url)

df = pd.read_csv(url,encoding="cp874")
df.head()
```

## EDA

### Remove column "TOTAL_DMG"

```python
df = df.drop('TOTAL_DMG', axis=1)
```

### Create column "TOTAL_DMG"

```python
df["TOTAL_DMG"] = df["UP_DMG"] + df["WALL_DMG"] + df["FLOOR_DMG"] + df["PILLAR_DMG"] + df["STRUCTURE_DMG"]
df.info()
```

### Remove District is na

```python
df = df.dropna(subset=['DISTRICT'])
```

### Remove Columns not related

```python
df = df.drop(["ZONE","SUB_ZONE","SoftClay_m","WaterConte","LiquidLimi","Plasticity","Total_Unit", "Score_Rang","Zoning", "ผังสี"], axis=1)
```

```python
df.groupby('DISTRICT')['TOTAL_DMG'].sum().sort_values(ascending=False)
```

```python
df['report_time'] = pd.to_timedelta(
    df['HOUR_'], unit='h'
) + pd.to_timedelta(df['MINUTE_'], unit='m')
```
