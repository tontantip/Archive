```python
import pandas as pd
import re

thai_months = {
    'มกราคม': 1, 'กุมภาพันธ์': 2, 'มีนาคม': 3, 'เมษายน': 4,
    'พฤษภาคม': 5, 'มิถุนายน': 6, 'กรกฎาคม': 7, 'สิงหาคม': 8,
    'กันยายน': 9, 'ตุลาคม': 10, 'พฤศจิกายน': 11, 'ธันวาคม': 12
}

def parse_thai_date(date_str):
    # Clean the string by removing 'วัน' and 'ที่' and normalizing spaces
    cleaned_str = date_str.replace('วัน', '').replace('ที่', '').strip()
    cleaned_str = re.sub(r'\s+', ' ', cleaned_str) # Replace multiple spaces with a single space

    parts = cleaned_str.split()

    day = None
    month_name = None
    year = None

    for part in parts:
        if part.isdigit():
            if len(part) == 4 and year is None: # Assume 4-digit number is the year
                year = int(part)
            elif len(part) <= 2 and day is None: # Assume 1 or 2-digit number is the day
                day = int(part)
        elif part in thai_months and month_name is None:
            month_name = part

    if day is not None and month_name is not None and year is not None:
        month = thai_months[month_name]
        return pd.Timestamp(year, month, day)
    else:
        raise ValueError(f"Could not parse date: {date_str}")

df['date_parsed'] = df['DATE_D'].apply(parse_thai_date)

# รวม date + hour → datetime รายชั่วโมง
df['datetime_hour'] = df['date_parsed'] + pd.to_timedelta(df['HOUR_'], unit='h')
```

## Modeling

```python
df['damage_level'] = pd.cut(
    df['TOTAL_DMG'],
    bins=[-1,0,2,5,10],
    labels=['none','minor','moderate','severe']
)
```

### set feature

```python
features = [
    'HOUR_',        # พฤติกรรมการแจ้ง
    'LAT','LON',
    'Zoning', 'ผังสี','Plasticity'
]
```

### Encode label

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

feature['Zoning'] = le.fit_transform(feature['Zoning'].astype(str))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

feature['ผังสี'] = le.fit_transform(feature['ผังสี'].astype(str))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

feature['Plasticity'] = le.fit_transform(feature['Plasticity'].astype(str))

```

### splite data

```python
from sklearn.model_selection import train_test_split

X = df[features]
y = df['damage_level']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,     # สำคัญมาก (class imbalance)
    random_state=42
)
```

### Train model Logistic regression

```python
from sklearn.linear_model import LogisticRegression

#model = LogisticRegression(max_iter=1000)
model = LogisticRegression(
    max_iter=1000
)
model.fit(X_train, y_train)
```

### Prediction use logistic regression model

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

### Train model Random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=300,
    max_depth=10,
    min_samples_leaf=10,
    max_features='sqrt',
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)
rf.fit(X_train, y_train)
```

### Prediction use random forest model

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
```

## Show F1 score

```python
from sklearn.metrics import f1_score

print("LogReg F1:", f1_score(y_test,model.predict(X_test), average='weighted'))
print("RF F1:", f1_score(y_test, rf.predict(X_test), average='weighted'))
```
